#### Scoring Matches

In this workflow, the spectrum for the sample will be score against each spectrum in the library. One approach is to score matches based on their "hit quality index" (HQI, typical score range between zero and one), which can depend on mathematical properties such as the Euclidean distance, the mean absolute deviation, or the correlation coefficient. High scoring matches are reported to the user. 

Many newer methods also contextualize this score by estimating a p-value. 
- A null distribution of HQI scores is created by comparing the spectrum for the target analyte with those spectra for random molecules that are not the target. These random comparisons establish what HQI scores would be expected by chance.
- For an observed HQI score, the p-value is calculated as the proportion of null distribution scores that are greater than or equal to the observed score.
- Critical HQI values can be determined for common significance levels (e.g., $\alpha = 0.05$)

In the example below, four samples are qualitatively analyzed to determine whether the sample is microcrystalline cellulose. Consistent with all samples being different cellulose containing species, they all yield HQI that are close to the upper limit of one. However, only the true microcrystalline cellulose sample passes the probability test at the 95% confidence level ($p-value > 0.05$)

**Table 2.** Sample data from Thermo-Scientific, testing the null hypothesis that the sample is microcrystalline cellulose (alternative hypothesis: the sample is not microcrystalline cellulose)
| Sample | p-value | HQI |
| --- | --- | --- |
| Microcrystalline cellulose | 0.338 | 0.9998 |
| Hydroxyethyl Cellulose | 0.00000754 | 0.9970 |
| Methyl Cellulose | 0.00000185 | 0.9766 |
| Hydroxypropyl Cellulose | 0.0000000323 |0.9796 |

Here, we described library searching and p-value estimation for FT-IR spectra, but this approach is also used for many of experiments that generate spectra with many features for individual analytes (e.g., mass spectrometry).

**Figure 3.** A Herriott cell is a multipass optical cavity consisting of two spherical mirrors of equal focal length set parallel to each other. Light enters through a hole in one mirror and undergoes multiple reflections between the mirrors in a systematic pattern before exiting through another hole. This design creates a long effective pathlength in a compact space, typically achieving 20-200 passes. ### Cavity Ring-Down Spectroscopy

Cavity Ring-Down Spectroscopy (CRDS) is a laser-based absorption technique that measures the decay rate of light trapped in a high-finesse optical cavity formed by two highly reflective mirrors (typically R > 99.99%). A laser pulse is injected into the cavity, and the time-dependent decay of the trapped light (the "ring-down") is measured. When a sample in the cavity absorbs light, it accelerates the decay rate. By comparing decay rates with and without the sample, the absolute absorption can be determined independent of laser power fluctuations. 

![Signal from cavity ring-down spectroscopy](https://www.iup.uni-bremen.de/troposphere/images/image001crds.jpg)

**Figure 4.** Schematic ring-down signals with (red curve) and without (black curve) absorbing medium. From the Tropospheric Radical Observations and Laser Absorption Spectroscopy research group at the University of Bremen.

The technique's high sensitivity comes from both the long effective pathlength (typically kilometers) created by multiple reflections and its ability to measure absolute time constants rather than relative changes in intensity.

### Attenuated Total Reflection

Attenuated Total Reflection (ATR) IR spectroscopy uses a high refractive index crystal (typically diamond, germanium, or zinc selenide) in contact with a sample. IR light enters the crystal at an angle greater than the critical angle, undergoes total internal reflection, and creates an evanescent wave that penetrates a few microns into the sample. The light is partially absorbed by the sample during each reflection, resulting in an IR spectrum. The technique requires minimal sample preparation, works well with aqueous solutions (due to short pathlength), and is especially useful for surface analysis, thick or strongly absorbing materials, and hard-to-prepare samples like polymers, pastes, powders, and liquids.

## Near-IR Spectroscopy

The near-IR region is primarily comprised of overtone and combination bands. These transitions typically have weaker molar absorptivities, but are particularly useful for routine quantitative analysis in industrial applications, especially for quality control and process monitoring.

Some benefits:
- Higher Signal-to-Noise Ratio:
    - Near-IR sources are more intense than mid-IR sources
    - Detectors have better performance in the near-IR region
    - Lower instrumental noise leads to better precision
- Weaker absorption allows direct measurement of concentrated samples, without dilution.
- These wavelength easily couple with fiber optic probes, which can be used for remote sampling

### Chemometrics

Previously, we introduced classical least squares as a method quantify the relative abundance of multiple components in a sample based on spectral vectors.

#### Classical Least Squares:
- Requires knowledge of all constituents' pure spectra
- Assumes Beer's Law holds (linear additivity)
- Directly solves for concentrations using known spectra
- Cannot handle spectral interactions or matrix effects
- More sensitive to baseline shifts and spectral artifacts
- Works best with simple, well-characterized systems

#### Partial Least Squares:
- Does not require knowledge of pure component spectra
- Uses both spectral and concentration data to build model
- Creates latent variables that maximize covariance between spectra and concentrations
- Can handle spectral interactions and matrix effects
- More robust to baseline variations and noise
- Better suited for complex mixtures and real-world samples
- Risk of overfitting if too many latent variables used

#### Principle Component Analysis

Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms high-dimensional data into a new set of uncorrelated variables called principal components (PCs). 
- Each PC is a linear combination of the original variables, constructed to capture the maximum remaining variance in the data. 
- The first PC accounts for the largest variance, the second PC captures the second most variance while being orthogonal to the first, and so on. 
- This transformation produces a new coordinate system where the PCs are ordered by their importance in explaining data variance. 
- PCA is commonly used for data compression, pattern recognition, and removal of collinearity in spectroscopic data
