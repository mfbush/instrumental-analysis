## 2-1
> For assembling the voltage divider shown in the book, two of each of the following resistors are available: $250 \Omega$, $500 \Omega$, and $1.00 K\Omega$.

**(a)** Propose a combination of resistors that would yield tha indicated voltages.

Using Kirchoff's voltage law:

$$ V_3 = 10.0 \, V - V_1 - V_2 = 5.0 \, V $$

Using the properties of a voltage divider:

$$ \frac{1.0}{10.0} = \frac{R_1}{R_1+R_2+R_3} $$

$$ \frac{4.0}{10.0} = \frac{R_2}{R_1+R_2+R_3} $$

$$ \frac{5.0}{10.0} = \frac{R_3}{R_1+R_2+R_3} $$

Combine Equations 2 and 3:

$$ \frac{R_2}{R_1} = \frac{4.0 \, V}{1.0 \, V} $$

This is true when $R_2 = 1 \, k\Omega$ and $R_1 = 250 \, \Omega$

Combine Equations 3 and 4:

$$ \frac{R_3}{R_2} = \frac{5.0 \, V}{4.0 \, V} = 1.25 $$

This is true when $R_3 = \, 1250 \Omega$, which can be achieved with a $1000 \, \Omega$ and a $250 \, \Omega$ resistor in series.

**(b)** *IR* drop across $R_3$. 

$$I \cdot R_3 = V_3 = 10.0 \, V - 1.0 \, V - 4.0 \, V = 5.0 \, V $$

**(c)** What current would be drawn from the source?

$$ I = \frac{V}{R_1 + R_2 + R_3} = \frac{10.0 \, V}{250 \, \Omega + 1000 \, \Omega + 1250 \, \Omega} = 0.004 \, A $$

**(d)** What power is dissipated by the circuit?

$$ P = I \cdot V = 0.004 \, A \cdot 10.0 \, V = 0.04 W $$

## 2-6
> For the circuit shown in the book, calculate:

The resistance between points 1 and 2:

$$R_{1,2} = \frac{3.0 \, K \Omega \cdot 4.0 \, K \Omega}{3.0 \, K \Omega + 4.0 \, K \Omega} = 1.7 \, K\Omega $$

The resistance between points 2 and 3:

$$R_{2,3} = \frac{2.0 \, K \Omega \cdot 1.0 \, K \Omega}{2.0 \, K \Omega + 1.0 \, K \Omega} = 0.7 \, K\Omega $$

Therefore:

$$ R_{total} = 1.7 \, K\Omega + 0.7 \, K\Omega + 2.0 \, K\Omega = 4.4 \, K\Omega $$

$$ I = \frac{24 \, V}{4.4 \, K\Omega} = 0.0055 \, A $$

> **(a)** The power dissipated between points 1 and 2.

$$ P_{1,2} = I^2 \cdot R_{1,2} = (0.0055 \, A)^2 \cdot 1.7 \, K\Omega = 0.052 \, W $$

> **(b)** The current drawn from the source.

From Kirchoff's current law and above:

$$ I = 0.0055 \, A $$

> **(c)** The voltage drop across resistor $R_A$.

$$ V_A = I \cdot R_A = 0.0055 \, A \cdot 2000 \, \Omega = 11.0 \, V $$

> **(d)** The voltage drop across resistor $R_D$.

$$ V_D = 24 \, V \cdot \frac{R_{2,3}}{R_{total}} = 3.65 \, V $$

> **(e)** The potential difference between points 5 and 4.

$$ V_{5,4} = 24.0 \, V - 11.0 \, V = 13.0 \, V $$

## 2-10
> The current in a circuit is to be determined by measuring the voltage drop across a precision resistor in series with the circuit.

> **(a)** What should be the resistance of the resistor in ohms if 1.00 V corresponds to 20 $\mu A$?

$$ R_s = \frac{V}{I} = \frac{1.00 \, V}{0.000020 \, A} = 50000 \, \Omega $$

> **(b)** What must be the resistance of the voltage-measuring device if the error in the current measurement is to be less than 2.0% relative.

Using Equation 2-19:

$$ Relative \, error = -2.0\% = \frac{50 \, k\Omega}{R_M + 50 \, k\Omega} $$

$$ 2500 \, k\Omega = R_M + 50 \, k\Omega $$

$$ R_M = 2450 \, K\Omega $$



## 2-14
> How long would it take to discharge a 0.025 $\mu F$ capacitor to 1% of its full charge through a resistance of:

$$ v_c = V_c \cdot e^{-\frac{t}{RC}} $$

For discharge to 1%:

$$ \frac{1.0}{100.0} = \frac{v_c}{V_c} = e^{-\frac{t}{R \cdot 25 nF}} $$

$$ ln(0.010) = -4.61 = \frac{t}{2.5 \cdot 10^{-8} \cdot R} $$

$$ t = 4.61 \cdot 2.5 \cdot 10^{-8} \cdot R = 1.15 \cdot 10^{-7} \cdot R $$

> **(a)** 10 $M\Omega$

$$ t = 1.15 \, seconds $$

> **(b)** 1 $M\Omega$

$$ t = 0.115 \, seconds $$

> **(c)** 1 $k\Omega$

$$ t = 1.15 \cdot 10^{-4} \, seconds $$
